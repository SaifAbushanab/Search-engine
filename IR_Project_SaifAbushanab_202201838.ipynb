```python
from nltk.stem import *
from nltk.stem.porter import *

import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
print(stopwords.words('english'))
```

```python
# 1) Clean

def clean(text):
   text = re.sub(r"http\S+", " ", str(text)) # remove urls
   text = re.sub(r"RT ", " ", str(text)) # remove rt
   text = re.sub(r"@[\w]*", " ", str(text)) # remove handles
   text = re.sub(r"[\.\,\#_\|\:\?\?\/\=]", " ", str(text)) # remove special characters
   text = re.sub(r'\t', ' ', str(text)) # remove tabs
   text = re.sub(r'\n', ' ', str(text)) # remove line jump
   text = re.sub(r"\s+", " ", str(text)) # remove extra white space
   text = str(text).strip()
   return str(text)


df["processed_text"] = df["Text"].apply(clean)
df
```

```python
# 2) Remove stop words

def remove_stop(text) :
  tokens = word_tokenize(str(text))
  sentence = []
  for i in tokens:
    if i not in stop_words:
      sentence.append(i)
  return' '.join(sentence)


df["processed_text"] = df["processed_text"].apply(remove_stop)
df
```

```python
# 3) Steeming

def steeming(text) :
  tokens = word_tokenize(text)
  steemed_text = []
  j = 0
  for i in tokens:
    steemed_text.append(stemmer.stem(i))
    #print(f"tokens : {steemed_text[j]}")
    j = j+1
  return ' '.join(steemed_text)


df["processed_text"] = df["processed_text"].apply(steeming)
df
```

# 3. Indexing:


```python
df['docno'] = df['docno'].astype(str)

indexer = pt.DFIndexer("./myFirstIndex", overwrite=True)
index_ref = indexer.index(df["processed_text"], df["docno"])
index = pt.IndexFactory.of(index_ref)
```

#4. Query Processing:

```python
def preprocess(word):
  word = remove_stop(word)
  word = steeming(word)
  return word
```

```python
query = "hello"
query = preprocess(query)
```

```python
# Iterate over the lexicon to find the entry for the term "hello"
for kv in index.getLexicon():
    if kv.getKey() == query:
        print("Key:", kv.getKey())
        print("Value:", kv.getValue().toString().split('@')[0])
        break
```

```python
# Initialize an empty list to store job IDs containing "hello"
docID = []

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Tokenize the term description and check for "hello" in the tokens
    tokens = row['processed_text'].lower().split()
    if "hello" in tokens:
        # If it contains "Hello", append the doc ID to the list
        docID.append(row['docno'])

# Count the number of occurrences of "Hello"
numDocID = len(docID)

# Print the count and the list of job IDs
print("Number of times the term 'hello' appeared:", numDocID)
print("Doc IDs:", docID)
```
